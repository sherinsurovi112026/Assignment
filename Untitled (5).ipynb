```python
#1
# =========================
# IMPORT LIBRARIES
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
import statsmodels.api as sm

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, roc_curve, auc

# =========================
# LOAD DATA
# =========================
df = pd.read_csv(r"C:\Users\ayash\OneDrive\Desktop\problem1\lung_disease.csv")

print("\nHEAD:\n", df.head())
print("\nINFO:\n", df.info())

# =========================
# ENCODING (IMPORTANT FIX)
# =========================
df["LungDisease"] = df["LungDisease"].map({"No": 0, "Yes": 1})
df["Smoking"] = df["Smoking"].map({"No": 0, "Yes": 1})
df["Pollution"] = df["Pollution"].map({"Low": 0, "High": 1})
df["Income"] = df["Income"].map({"Low": 0, "Medium": 1, "High": 2})

# =========================
# EDA
# =========================

# Age distribution
sns.histplot(df["Age"], kde=True)
plt.title("Age Distribution")
plt.show()

print("\nSkewness of Age:", df["Age"].skew())

# Proportions
print("\nSmoking:\n", df["Smoking"].value_counts(normalize=True))
print("\nPollution:\n", df["Pollution"].value_counts(normalize=True))
print("\nLung Disease:\n", df["LungDisease"].value_counts(normalize=True))

# =========================
# CROSSTAB
# =========================
ct_smoke = pd.crosstab(df["Smoking"], df["LungDisease"])
ct_poll = pd.crosstab(df["Pollution"], df["LungDisease"])

print("\nSmoking vs Lung Disease:\n", ct_smoke)
print("\nPollution vs Lung Disease:\n", ct_poll)

print("\nRow % Smoking:\n", ct_smoke.div(ct_smoke.sum(axis=1), axis=0))
print("\nRow % Pollution:\n", ct_poll.div(ct_poll.sum(axis=1), axis=0))

# =========================
# CHI-SQUARE TEST
# =========================
chi2, p1, _, _ = stats.chi2_contingency(ct_smoke)
print("\nChi-square Smoking p-value:", p1)

chi2, p2, _, _ = stats.chi2_contingency(ct_poll)
print("Chi-square Pollution p-value:", p2)

# =========================
# ODDS RATIO
# =========================
if ct_smoke.shape == (2,2):
    OR = (ct_smoke.iloc[1,1]*ct_smoke.iloc[0,0]) / (ct_smoke.iloc[0,1]*ct_smoke.iloc[1,0])
    print("\nOdds Ratio (Smoking):", OR)

# =========================
# LOGISTIC REGRESSION (STATSMODELS)
# =========================
X = df[["Smoking", "Age", "Pollution", "Income"]]
X = sm.add_constant(X)
y = df["LungDisease"]

model = sm.Logit(y, X).fit()

print("\nLOGISTIC REGRESSION SUMMARY:\n", model.summary())
print("\nOdds Ratios:\n", np.exp(model.params))

# =========================
# INTERACTION MODEL
# =========================
df["Smoking_Pollution"] = df["Smoking"] * df["Pollution"]

X2 = df[["Smoking", "Age", "Pollution", "Income", "Smoking_Pollution"]]
X2 = sm.add_constant(X2)

model2 = sm.Logit(y, X2).fit()

print("\nINTERACTION MODEL:\n", model2.summary())

# =========================
# ML LOGISTIC REGRESSION
# =========================
X_ml = df[["Smoking", "Age", "Pollution", "Income"]]
y_ml = df["LungDisease"]

X_train, X_test, y_train, y_test = train_test_split(
    X_ml, y_ml, test_size=0.3, random_state=42
)

lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)

y_pred = lr.predict(X_test)
y_prob = lr.predict_proba(X_test)[:,1]

# =========================
# EVALUATION
# =========================
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nAccuracy:", accuracy_score(y_test, y_pred))

fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f"AUC = {roc_auc:.2f}")
plt.plot([0,1],[0,1],'--')
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC Curve")
plt.legend()
plt.show()
```

```python
#2
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd

# -----------------------------
# Example dataset (replace with your actual data)
# -----------------------------
np.random.seed(42)

# Simulated weight loss data (in pounds)
program_A = np.random.normal(loc=5.2, scale=1.1, size=30)
program_B = np.random.normal(loc=6.8, scale=1.3, size=30)
program_C = np.random.normal(loc=4.9, scale=1.0, size=30)

# Combine into a DataFrame
df = pd.DataFrame({
    'weight_loss': np.concatenate([program_A, program_B, program_C]),
    'program': ['A']*30 + ['B']*30 + ['C']*30
})

# -----------------------------
# Step 1: One-way ANOVA
# -----------------------------
model = ols('weight_loss ~ C(program)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
print("ANOVA Table:\n", anova_table)

# -----------------------------
# Step 2: Assumption Checks
# -----------------------------
# Normality check (Shapiro-Wilk test)
print("\nShapiro-Wilk Test (Normality):")
for prog in ['A','B','C']:
    stat, p = stats.shapiro(df[df['program']==prog]['weight_loss'])
    print(f"Program {prog}: W={stat:.3f}, p={p:.3f}")

# Homogeneity of variances (Levene’s test)
levene_stat, levene_p = stats.levene(program_A, program_B, program_C)
print(f"\nLevene’s Test: Stat={levene_stat:.3f}, p={levene_p:.3f}")

# -----------------------------
# Step 3: Post-hoc Analysis (Tukey HSD)
# -----------------------------
tukey = pairwise_tukeyhsd(endog=df['weight_loss'], groups=df['program'], alpha=0.05)
print("\nTukey HSD Results:\n", tukey)
```

```python
# =========================================================
# Chi-Square Test of Association
# Children Anemia Dataset
# =========================================================

# =========================================================
# Step 1: Import Libraries
# =========================================================

import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from scipy.stats import chi2_contingency

import os

# =========================================================
# Step 2: Create Output Folder
# =========================================================

save_path = (
    r"C:\Users\ayash\OneDrive"
    r"\Desktop\problem3\output"
)

os.makedirs(
    save_path,
    exist_ok=True
)

# =========================================================
# Step 3: Load Dataset
# =========================================================

df = pd.read_csv(
    r"C:\Users\ayash\OneDrive"
    r"\Desktop\problem3\children anemia.csv"
)

# =========================================================
# Step 4: Preview Dataset
# =========================================================

print("\nFirst 5 Rows:\n")

print(df.head())

print("\nDataset Information:\n")

print(df.info())

# =========================================================
# Step 5: Contingency Table
# =========================================================

table = pd.crosstab(

    df['Age in 5-year groups'],

    df['Anemia level']

)

print(
    "\n=============== "
    "CONTINGENCY TABLE "
    "===============\n"
)

print(table)

# =========================================================
# Step 6: Chi-Square Test
# =========================================================

chi2, p, dof, expected = (

    chi2_contingency(table)

)

print(
    "\n=============== "
    "CHI-SQUARE TEST RESULTS "
    "===============\n"
)

print(f"Chi-square Statistic : {chi2:.4f}")

print(f"p-value              : {p:.6f}")

print(f"Degrees of Freedom   : {dof}")

# =========================================================
# Step 7: Expected Frequencies
# =========================================================

expected_df = pd.DataFrame(

    expected,

    index=table.index,

    columns=table.columns

)

print(
    "\n=============== "
    "EXPECTED FREQUENCIES "
    "===============\n"
)

print(expected_df)

# =========================================================
# Step 8: Save Tables
# =========================================================

table.to_csv(
    save_path +
    r"\contingency_table.csv"
)

expected_df.to_csv(
    save_path +
    r"\expected_frequencies.csv"
)

# =========================================================
# Step 9: Contribution Calculation
# =========================================================

contribution = (
    (table - expected) ** 2
) / expected

print(
    "\n=============== "
    "CHI-SQUARE CONTRIBUTIONS "
    "===============\n"
)

print(contribution)

# =========================================================
# Step 10: Heatmap Visualization
# =========================================================

plt.figure(figsize=(10, 6))

sns.heatmap(

    contribution,

    annot=True,

    cmap="coolwarm",

    fmt=".2f"

)

plt.title(
    "Chi-square Contribution Heatmap"
)

plt.xlabel("Anemia Level")

plt.ylabel("Age Group")

# Save figure
plt.savefig(

    save_path +
    r"\chi_square_contribution.png",

    dpi=300,

    bbox_inches="tight"

)

plt.show()

# =========================================================
# Step 11: Interpretation
# =========================================================

print(
    "\n=============== "
    "INTERPRETATION "
    "===============\n"
)

if p < 0.05:

    print(
        "There is a statistically "
        "significant association "
        "between age group and "
        "anemia level."
    )

else:

    print(
        "There is NO statistically "
        "significant association "
        "between age group and "
        "anemia level."
    )

# =========================================================
# Step 12: Save Results to Text File
# =========================================================

with open(
    save_path + r"\results.txt",
    "w"
) as f:

    f.write(
        "Chi-square Test Results\n"
    )

    f.write(
        "------------------------\n"
    )

    f.write(
        f"Chi-square: {chi2}\n"
    )

    f.write(
        f"p-value: {p}\n"
    )

    f.write(
        f"Degrees of Freedom: {dof}\n\n"
    )

    if p < 0.05:

        f.write(
            "Conclusion: Significant "
            "association exists.\n"
        )

    else:

        f.write(
            "Conclusion: No significant "
            "association.\n"
        )

# =========================================================
# Step 13: Final Output Message
# =========================================================

print(
    "\nAll outputs saved in:\n"
)

print(save_path)
```

```python
#4
import pandas as pd
import numpy as np
from scipy.stats import fisher_exact
from statsmodels.stats.multitest import multipletests
from itertools import combinations

# =====================================================
# Create Contingency Table
# =====================================================

data = np.array([
    [40, 10],   # Drug A
    [10, 40],   # Drug B
    [25, 25]    # Drug C
])

df = pd.DataFrame(
    data,
    index=['Drug A', 'Drug B', 'Drug C'],
    columns=['No Disease', 'Disease']
)

print("Contingency Table:\n")
print(df)

# =====================================================
# Pairwise Post-hoc Fisher Tests
# =====================================================

pairs = list(combinations(df.index, 2))

p_values = []
results = []

for pair in pairs:

    sub_table = df.loc[list(pair)]

    oddsratio, p = fisher_exact(sub_table)

    p_values.append(p)

    results.append([
        pair[0],
        pair[1],
        round(oddsratio, 3),
        p
    ])

# =====================================================
# Bonferroni Correction
# =====================================================

adjusted = multipletests(
    p_values,
    method='bonferroni'
)

# =====================================================
# Display Results
# =====================================================

print("\nPost-hoc Fisher Exact Test Results:\n")

for i, res in enumerate(results):

    print("Comparison :", res[0], "vs", res[1])
    print("Odds Ratio :", res[2])
    print("Raw p-value :", round(res[3], 5))
    print("Adjusted p-value :",
          round(adjusted[1][i], 5))

    if adjusted[1][i] < 0.05:
        print("Result : Significant Difference\n")
    else:
        print("Result : No Significant Difference\n")
```

```python
#6
# =========================================================
# Poisson Regression (GLM)
# Number of Awards Earned by High School Students
# =========================================================

# Step 1: Import required libraries
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf

# =========================================================
# Step 2: Load dataset from URL
# =========================================================
url = "https://stats.idre.ucla.edu/stat/data/poisson_sim.csv"

df = pd.read_csv(url)

# =========================================================
# Step 3: Display dataset information
# =========================================================
print("\nFirst 5 Rows:\n")
print(df.head())

print("\nDataset Information:\n")
print(df.info())

print("\nSummary Statistics:\n")
print(df.describe())

# =========================================================
# Step 4: Convert 'prog' into categorical variable
# =========================================================
df["prog"] = df["prog"].astype("category")

# =========================================================
# Step 5: Fit Poisson Regression Model
# =========================================================
model = smf.glm(
    formula = "num_awards ~ math + prog",
    data = df,
    family = sm.families.Poisson()
).fit()

# =========================================================
# Step 6: Model Summary
# =========================================================
print("\n================ POISSON REGRESSION SUMMARY ================\n")
print(model.summary())

# =========================================================
# Step 7: Incidence Rate Ratios (IRR)
# =========================================================
irr = np.exp(model.params)

print("\n================ INCIDENCE RATE RATIOS (IRR) ================\n")
print(irr)

# =========================================================
# Step 8: Predicted Counts
# =========================================================
df["Predicted_Count"] = model.predict(df)

print("\n================ PREDICTED NUMBER OF AWARDS ================\n")
print(df[["num_awards", "Predicted_Count"]].head(10))

# =========================================================
# Step 9: Interpretation
# =========================================================
print("\n================ INTERPRETATION ================\n")

for variable, pvalue in model.pvalues.items():

    if pvalue < 0.05:
        significance = "Statistically Significant"
    else:
        significance = "Not Significant"

    print(f"{variable} --> p-value = {pvalue:.4f} --> {significance}")

print("\nConclusion:")
print("""
1. Math score has a positive effect on the number of awards earned.
2. Students in different academic programs earn different numbers of awards.
3. Significant predictors (p < 0.05) strongly influence award counts.
4. IRR > 1 indicates increased expected award counts.
5. IRR < 1 indicates decreased expected award counts.
""")
```

```python
#10
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.optimize import minimize
from scipy.special import gammaln

# -----------------------------
# 1. CREATE SYNTHETIC DATA (ZTP STYLE)
# -----------------------------
np.random.seed(123)

n = 300

age = np.random.randint(20, 80, n)
insurance = np.random.choice([0, 1, 2], n)   # 0=none,1=govt,2=private
death = np.random.choice([0, 1], n)

X = np.column_stack([age, insurance, death])
X = sm.add_constant(X)

beta_true = np.array([0.2, 0.01, -0.15, 0.5])

mu = np.exp(X @ beta_true)

# Poisson generation
y = np.random.poisson(mu)

# truncate zeros (ZERO-TRUNCATED DATA)
y = y[y > 0]
X = X[:len(y)]

# -----------------------------
# 2. ZTP LOG-LIKELIHOOD
# -----------------------------
def ztp_loglike(params, y, X):
    eta = X @ params
    mu = np.exp(eta)

    loglik = y * np.log(mu) - mu - gammaln(y + 1)
    loglik -= np.log(1 - np.exp(-mu))

    return -np.sum(loglik)

# -----------------------------
# 3. OPTIMIZATION
# -----------------------------
init = np.zeros(X.shape[1])

res = minimize(ztp_loglike, init, args=(y, X), method="BFGS")

# -----------------------------
# 4. RESULTS
# -----------------------------
print("Coefficients:")
print(res.x)

print("\nIRR:")
print(np.exp(res.x))
```

```python
#11
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.optimize import minimize
from scipy.special import gammaln

# -----------------------------
# 1. SIMULATE HOSPITAL DATA
# -----------------------------
np.random.seed(42)

n = 400

age = np.random.randint(18, 90, n)
insurance = np.random.choice([0, 1, 2], n)  # 0=no,1=govt,2=private
death = np.random.choice([0, 1], n)

X = np.column_stack([age, insurance, death])
X = sm.add_constant(X)

beta = np.array([0.3, 0.015, -0.25, 0.8])

mu = np.exp(X @ beta)

# generate counts
y = np.random.negative_binomial(1, 1/(1+mu))

# zero-truncate (hospital stay >= 1 day)
mask = y > 0
y = y[mask]
X = X[mask]

# -----------------------------
# 2. ZT NEGATIVE BINOMIAL MODEL
# -----------------------------
def ztnb_loglike(params, y, X, alpha=1.0):
    eta = X @ params
    mu = np.exp(eta)

    loglik = (
        gammaln(y + 1/alpha)
        - gammaln(y + 1)
        - gammaln(1/alpha)
        + y*np.log(mu)
        - (y + 1/alpha)*np.log(1 + alpha*mu)
    )

    p0 = (1 + alpha*mu)**(-1/alpha)
    loglik -= np.log(1 - p0)

    return -np.sum(loglik)

# -----------------------------
# 3. FIT MODEL
# -----------------------------
init = np.zeros(X.shape[1])

res = minimize(ztnb_loglike, init, args=(y, X), method="BFGS")

# -----------------------------
# 4. RESULTS
# -----------------------------
print("Coefficients:")
print(res.x)

print("\nIRR:")
print(np.exp(res.x))
```
